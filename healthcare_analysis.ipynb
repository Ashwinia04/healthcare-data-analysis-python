```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
diabetes.info()
```

```python
diabetes.describe
```

```python
diabetes['Outcome'].value_counts().plot(kind='bar')

plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Count")

plt.show()
```

```python
plt.hist(diabetes['Age'])

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

## Insights
- The dataset contains patient healthcare information related to diabetes.
- Outcome distribution shows diabetic and non-diabetic cases.
- Age distribution indicates higher representation of middle-aged individuals.
